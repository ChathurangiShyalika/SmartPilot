```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import pmdarima as pm
from statsmodels.tsa.seasonal import seasonal_decompose

# Read CSV file
production_df = pd.read_csv("run.csv")
production_df.columns = ['id', 'comment', 'whenStarted', 'whenEnded', 'Process', 'Product']

# Group products based on the time process ends
demand_per_time = production_df.groupby(['whenEnded', 'Process', 'Product'])['whenEnded'].count().reset_index(name="produced")
demand_per_time.rename(columns={'whenEnded': 'timestamp'}, inplace=True)

# Convert string timestamp to pandas datetime
demand_per_time['timestamp'] = demand_per_time['timestamp'].astype('datetime64[ns]')
demand_per_time.head()

# Remove time from timestamp and keep the date
demand_per_time['timestamp'] = demand_per_time['timestamp'].dt.strftime('%Y-%m-%d %H:00:00')
demand_per_time.head()

# Get number of products per hour
demand_per_h = demand_per_time.groupby(['timestamp', 'Process', 'Product'])['timestamp'].count().reset_index(name="produced")
demand_per_h.head()

# Save the edited DataFrame to CSV
demand_per_h.to_csv("run-edited-h.csv", index=False)

# Set the date column as the index
demand_per_h.set_index('timestamp', inplace=True)
demand_per_h.head()

# Plot the data
plt.figure(figsize=(20, 5))
demand_per_h.groupby('Product')['produced'].plot(legend=True)
plt.show()

file_path = "run-edited-h.csv"
data = pd.read_csv(file_path)

# Convert date column to datetime if not already
data['timestamp'] = pd.to_datetime(data['timestamp'])

# Set the date column as the index
data.set_index('timestamp', inplace=True)

# Ensure data is sorted by date
data.sort_index(inplace=True)

# List of unique product types
product_types = data['Product'].unique()

# Dictionary to store models for each product
models = {}

# Dictionary to store results DataFrames for each product
results = {}

# Build, fit, and forecast using Auto-ARIMA for each product
for product in product_types:
    product_data = data[data['Product'] == product]['produced']
    
    # Perform seasonal decomposition
    decomposition = seasonal_decompose(product_data, model='additive', period=24)
    seasonal = decomposition.seasonal
    trend = decomposition.trend
    resid = decomposition.resid
    
    # Fill missing values in trend and resid with interpolated values
    trend = trend.interpolate()
    resid = resid.interpolate()
    
    # Difference the series to make it stationary
    product_data_diff = product_data.diff().dropna()
    
    train_size = int(len(product_data_diff) * 0.8)
    
    train = product_data_diff[:train_size]
    test = product_data_diff[train_size:]
    
    # Fit Auto-ARIMA model with fine-tuned parameters
    model = pm.auto_arima(train,
                          start_p=1, start_q=1,
                          max_p=5, max_q=5,
                          seasonal=True,
                          m=24,
                          start_P=0, seasonal_test='ocsb',
                          d=None, D=1, trace=True,
                          error_action='ignore',
                          suppress_warnings=True,
                          stepwise=True)
    models[product] = model

    # Forecast
    forecast_steps = len(test)
    forecast, conf_int = model.predict(n_periods=forecast_steps, return_conf_int=True)
    
    # Revert the differencing to get the actual forecast values
    forecast = pd.Series(forecast, index=test.index).cumsum() + product_data[train_size]
    
    # Adjust forecast length to match the test set
    forecast = forecast[:len(test)]
    
    # Evaluate the model
    mse = mean_squared_error(product_data[train_size+1:], forecast)
    print(f'{product} MSE: {mse}')

    # Create a DataFrame with actual values, forecasted values, and differences
    result_df = pd.DataFrame({
        'Actual': product_data[train_size+1:train_size+1+len(forecast)],
        'Forecasted': forecast,
        'Difference': product_data[train_size+1:train_size+1+len(forecast)] - forecast
    }, index=test.index[:len(forecast)])

    results[product] = result_df

    # Save the DataFrame to a CSV file
    result_df.to_csv(f'{product}_forecast_results.csv')

    # Plot the results
    plt.figure(figsize=(14, 7))
    plt.plot(product_data.index, product_data, label='Actual')
    plt.plot(forecast.index, forecast, label='Forecast')
    plt.fill_between(test.index[:len(forecast)], conf_int[:len(forecast), 0], conf_int[:len(forecast), 1], color='gray', alpha=0.2)
    plt.title(f'{product} - Auto-ARIMA Forecast')
    plt.xlabel('Date')
    plt.ylabel('Produced')
    plt.legend()
    plt.show()

# Print the results
for product, result_df in results.items():
    print(f"Product: {product}")
    print(result_df.head(10))
```
